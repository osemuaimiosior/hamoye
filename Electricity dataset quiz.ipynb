```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
url ='https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv?raw=true'
```

```python
df = pd.read_csv(url,error_bad_lines = False)
```

```python
data = df.copy()
```

```python
data.head()
```

## Lowest average fuel cost burned by fuel type

```python
average_fuel_cost = data.groupby(['fuel_type_code_pudl']).mean()
```

```python
average_fuel_cost
```

## Finding avearage year maximum fuel_cost_per_unit_delivered

```python
year = data.groupby(['report_year']).mean()
```

## Finding percentage change

```python
year_1994 = data[data['report_year'] == 1994]
```

```python
coal_1994= year_1994[year_1994['fuel_type_code_pudl'] == 'coal']
```

```python
coal_1994.head()
```

```python
pct_change_1994 = coal_1994.fuel_cost_per_unit_burned.iloc[0]-coal_1994.fuel_cost_per_unit_burned.iloc[-1]
```

```python
print(pct_change_1994)
```

```python
year_1998 = data[data['report_year'] == 1998]
year_1998.head()
```

```python
coal_1998= year_1998[year_1998['fuel_type_code_pudl'] == 'coal']
```

```python
pct_change_1998 = coal_1998.fuel_cost_per_unit_burned.iloc[0]-coal_1998.fuel_cost_per_unit_burned.iloc[-1]
pct_change_1998
```

```python
pct_change = pct_change_1998/pct_change_1994*100
print(pct_change)
```
